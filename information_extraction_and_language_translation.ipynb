```python
from openai import OpenAI
from IPython.display import display,Markdown,HTML
```

```python
f = open('keys/.openai_api_key.txt')
OPENAI_API_KEY=f.read()
client=OpenAI(api_key=OPENAI_API_KEY)
```

```python
clinical_prescription =f"""
CityCare Multispeciality Hospital, Pune, Maharashtra
Contact: +91-XXXXXXXXXX

OPD Prescription

Patient Name: Rahul Sharma
Age/Sex: 29 Years / Male
Date: 24/03/2026
OPD No.: CCMH/OPD/2026/4587

Chief Complaints: Patient reports fever for 2 days, associated with headache and generalized body ache. No history of cough, breathlessness, vomiting, or loose motions.

History of Present Illness: The fever is moderate-grade, intermittent, and partially relieved with over-the-counter medication. No recent travel or known exposure to infectious diseases reported.

Past History: No known chronic illnesses such as hypertension, diabetes, or asthma. No prior hospitalizations.

Family History: Father is a known case of hypertension. Mother is a known case of type 2 diabetes mellitus. No significant history of hereditary or infectious diseases in the family.

General Examination: Temperature 101°F, Pulse 88 bpm, Blood Pressure 118/76 mmHg, SpO₂ 98% on room air.

Systemic Examination: Cardiovascular system is normal. Respiratory system shows clear breath sounds with no added sounds. Abdomen is soft and non-tender. Central nervous system is normal, patient is conscious and oriented.

Provisional Diagnosis: Acute febrile illness, likely viral in origin.

Prescription: Paracetamol 500 mg to be taken every 6 hours as needed for fever, not exceeding 4 tablets per day. Vitamin C 500 mg once daily for supportive care. The patient is advised to maintain adequate hydration and take a light, nutritious diet.

Advice: Adequate rest is recommended. Monitor body temperature regularly. If fever persists beyond 3 days or symptoms worsen, seek medical attention immediately.

Investigations: Complete blood count, malaria parasite test, and dengue NS1 antigen test if symptoms persist.

Follow-Up: Review after 3 days or earlier if required.

Doctor’s Name & Signature: Dr. __________ (MBBS, MD General Medicine)

"""
```

```python
def get_completion(prompt,model="gpt-4o-mini"):
    message=[{"role":"user","content":prompt}]
    response=client.chat.completions.create(
        model=model,
        messages=message,
        temperature=0	
    )
    return response.choices[0].message.content
                   
    
```

```python
prompt=f"""
Act as an expert in analyzing and understanding clinical doctor notes in healthcare.
Extract all symptoms only from the clinical_prescription infromation below.
Difference between symptoms that are present vs absent
Give me probability (high/medium/low) of how sure you are about the result
Add note on the probabilities and why you think so.

output as a markdown table with the following columns,
all symptoms should be expanded and no acronyms unless you dont know.

Symptoms| Present/Denies | Probability 

also expand all acronyms.

Output that also as a seperate appendix table in markdown.
if no appendix is there in the clinical_prescription do not generate the table

Clinical Note =
{clinical_prescription}
"""
```

```python
response=get_completion(prompt)
display(Markdown(response))
```
